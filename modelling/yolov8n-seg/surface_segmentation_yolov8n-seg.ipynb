```python
from ultralytics import YOLO

from IPython.display import display, Image

model = YOLO('yolov8n-seg.pt')
```

```python
model.train(data='/content/drive/MyDrive/KITA_1026/project/ocr/data_surface_complete/data.yaml', epochs=50, patience=20, batch=32, imgsz=1024)
```
